# Create Model 


```python
x = data.drop('Outcome',axis=1)
y = data['Outcome']
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=.2) # splitng data to 4 parts
```

```python
model1=LogisticRegression()
model2=SVC()
model3=RandomForestClassifier()
model4=GradientBoostingClassifier(n_estimators=1000) # it's contain alot of a week algorithm we chooce some of them to work with other 
```

```python
columns=['LogisticRegression','SVC','RandomForestClassifier','GradientBoostingClassifier']
res_accurac=[]
res_recall=[]
res_f1=[]
```

```python
def cal(model):
    model.fit(x_train,y_train) # to train in x and y
    pre=model1.predict(x_test) # to prediction x
    accuracy=accuracy_score(pre,y_test) # accurace measurment
    recall=recall_score(pre,y_test) # performance measurment
    f1=f1_score(pre,y_test)
    #confusion_matrix(pre,y_test) #[[true_positive,true_negative],[false_positive,false_positive]]
    
    res_accurac.append(accuracy)
    res_recall.append(recall)
    res_f1.append(f1)
    
    sns.heatmap(confusion_matrix(pre,y_test),annot=True)
    print(model)
    print('accuracy is: ',accuracy,'recall i :',recall,'f1 is : ',f1)

cal(model1)
```

```python
cal(model2)
```

```python
cal(model3)
```

```python
cal(model4)
```
