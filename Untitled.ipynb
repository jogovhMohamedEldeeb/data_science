# Create Model

```python
x = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
model_1 = LogisticRegression()
model_2 = SVC()
model_3 = RandomForestClassifier()

model_4 = GradientBoostingClassifier(n_estimators=1000)# it's contain alot of a weqk algorithm we chooce some of them to work with other

```

```python
columns = ['LogisticRegression', 'SVC',
           'RandomForestClassifier', 'GradientBoostingClassifier']
res1 = []
res2 = []
res3 = []
```

```python
def call_model(model):
    model.fit(x_train,y_train) # to train in x and y
    pre=model.predict(x_test) # to prediction x
    accuracy=accuracy_score(pre,y_test) # accurace measurment
    recall=recall_score(pre,y_test) # performance measurment
    f1=f1_score(pre,y_test)
    #confusion_matrix(pre,y_test) #[[true_positive,true_negative],[false_positive,false_positive]]
    
    res1.append(accuracy)
    res2.append(recall)
    res3.append(f1)
    
    sns.heatmap(confusion_matrix(pre,y_test),annot=True)
    print(model)
    print('accuracy is: ',accuracy,'recall i :',recall,'f1 is : ',f1)
```

```python
call_model(model_1)
```

```python
call_model(model_2)
```

```python
call_model(model_3)
```

```python
call_model(model_4)
```
